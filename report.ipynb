# Titanic Dataset Feature Selection Analysis Report

This report uses a series of bar charts to illustrate the relationships between various features and passenger survival on the Titanic. The objective is to identify the most relevant features for survival prediction. (Figures generated from modularisation.ipynb)

## 1. Sex
   - The chart shows a significantly higher survival rate for female passengers compared to male passengers. Over 50% of females survived, while the survival rate for males is much lower. This aligns with the "women and children first" rescue protocol, making gender a critical factor in survival prediction.
   - **Feature Selection Decision**: Sex is a crucial feature and should be included in the survival prediction model.

## 2. Passenger Class (Pclass)
   - The data reveals that first-class passengers had the highest survival rate, while third-class passengers had the lowest. A large number of third-class passengers perished, which reflects the social priority differences during rescue.
   - **Feature Selection Decision**: Pclass should be included as it captures the relationship between social class and survival rate.

## 3. Age Interval
   - The analysis of age intervals shows that younger passengers (especially children) had a higher survival rate, reflecting priority in rescue efforts. Survival rates gradually decrease with age.
   - **Feature Selection Decision**: Age, particularly segmented into intervals, is a significant feature that can indicate survival likelihood.

## 4. Fare Interval
   - The chart indicates a slightly higher survival rate among passengers who paid higher fares. This trend could be due to higher-paying passengers generally being in first class, with better access to lifeboats and resources.
   - **Feature Selection Decision**: Fare interval should be included to capture the impact of economic status on survival.

## 5. Family Type
   - The chart reveals that single passengers have a lower survival rate, whereas those with smaller family groups have a higher chance of survival. Small families may have been more likely to assist each other during evacuation.
   - **Feature Selection Decision**: Family type is a valuable feature, reflecting how family structure impacts survival chances.

## 6. Titles
   - Different titles correlate with varied survival rates. For instance, passengers with titles like "Miss" and "Master" show higher survival rates, whereas "Mr." has the lowest survival rate. Titles provide additional social status and age information.

## Conclusion
Through these visual analyses, we identify sex, passenger class, age, fare interval, family type, and titles as key features for survival prediction. These features reflect rescue priority, social status, and family structure, each of which has a significant impact on survival. Including these features in the survival prediction model can improve its accuracy.
